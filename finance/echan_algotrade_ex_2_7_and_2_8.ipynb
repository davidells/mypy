### Example 2.7 from Ernie Chan's *Algorithmic Trading*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime

%matplotlib inline
```

```python
start_dt = datetime.datetime(2006, 4, 26)
end_dt = datetime.datetime(2012, 4, 9)
```

```python
import pandas.io.data as web
EWA = web.DataReader("EWA", 'yahoo', start_dt, end_dt)
EWC = web.DataReader("EWC", 'yahoo', start_dt, end_dt)
IGE = web.DataReader("IGE", 'yahoo', start_dt, end_dt)
```

```python
y3 = pd.concat(
    [EWC['Adj Close'], EWA['Adj Close'], IGE['Adj Close']], 
    axis=1, keys=['EWC', 'EWA', 'IGE'])
```

```python
from johansen.johansen import coint_johansen
```

```python
jo = coint_johansen(y3, 0, 1)
```

```python
jo.lr1
```

```python
jo.cvt
```

```python
jo.lr2
```

```python
jo.cvm
```

```python
jo.evec
```

```python
weights = jo.evec[:,0]
yport = (weights * y3).sum(axis=1)
```

```python
import mean_reversion as mr
mr.halflife(yport)
```

```python
import statsmodels.tsa.stattools as sms
sms.adfuller(yport)
```

```python
yport.plot(figsize=(16,9))
```

### Example 2.8

```python
lookback = round(mr.halflife(yport)) # Note: this is 21 due to different johansen result, boosts results below
zScore = (yport - pd.rolling_mean(yport, lookback)) / pd.rolling_std(yport, lookback)
positions = pd.concat([-zScore] * y3.shape[1], axis=1).values * weights * y3
pnl = (positions.shift(1) * y3.pct_change()).sum(axis=1)
grossMktVal = positions.shift(1).abs().sum(axis=1)
ret = pnl / grossMktVal
```

```python
((1 + ret).cumprod() - 1).plot(figsize=(10,8))
```

```python
sharpe = np.sqrt(252) * ret.mean() / ret.std()
apr = pow((1 + ret.mean()), 252) - 1
print "Sharpe ratio: %s, APR: %s" % (sharpe, apr)
```
